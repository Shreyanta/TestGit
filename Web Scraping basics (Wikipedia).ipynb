# Web Scraping tutorial for different sections of Wikipedia

Task 1 : Header


```python
import requests
from bs4 import BeautifulSoup
URL='https://en.wikipedia.org/wiki/Machine_learning'
response = requests.get(URL)
soup = BeautifulSoup(response.content, 'html.parser')
title = soup.find(id="firstHeading")
print(title.string)
```

Task 2: Content

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
URL='https://en.wikipedia.org/wiki/Machine_learning'
response = requests.get(URL)
soup = BeautifulSoup(response.content, 'html.parser')
results = soup.find(id="content")
titles = results.find_all('p')
for i in titles:
 print(i.getText())
```

Task 3: Table of Contents

```python
import requests
from bs4 import BeautifulSoup
URL = 'https://en.wikipedia.org/wiki/Machine_learning'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
results = soup.find(id='toc')
titles = results.find_all('a')
titles
ML=[]
for link in titles:
 ML.append(link.get('href'))
import pandas as pd
df=pd.DataFrame()
df['Contents']=ML
df
```

Task 4: References for a single page

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
URL = 'https://en.wikipedia.org/wiki/Machine_learning'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
title = soup.find(id="References")
print(title.string)
results = soup.find(class_='references')
titles = results.find_all('li')
for i in titles:
 print(i.getText())

    
```

Task 5: Counting the number of references in Multiple pages


```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
URL = ['https://en.wikipedia.org/wiki/Machine_learning','https://en.wikipedia.org/wiki/Web_scraping',
'https://en.wikipedia.org/wiki/Artificial_intelligence','https://en.wikipedia.org/wiki/Data_science']
for link in URL:
 page = requests.get(link)
 soup = BeautifulSoup(page.content, 'html.parser')
 results = soup.find(class_='references')
 titles = results.find_all('li')
 titles
 print(len(titles))
```
